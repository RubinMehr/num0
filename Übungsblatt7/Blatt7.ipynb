```python
import math
import numpy as np

def createX(N): #äquidistante Intervallzerlegung x_i = i/N (i=0,...,N)
    x=np.zeros(N+1)
    for i in range(len(x)):
        x[i]=i/N
    return x

def f(x):
    return 1/(1+2*x)

def trapez(N,x):
    summe = 0
    for i in range(N): #i=0,...,N-1
        summe += ((x[i+1]-x[i])/2)*(f(x[i])+f(x[i+1]))
    return summe

def simpson(N,x):
    summe = 0
    for i in range(N):
        summe += ((x[i+1]-x[i])/6)*(f(x[i])+4*f((x[i]+x[i+1])/2)+f(x[i+1]))
    return summe


for N in range(2,102,2): #Beginnt mit N=2, eerhöht in 2er Schritten bis maximal N=100
    x = createX(N)
    It = trapez(N,x)
    Is = simpson(N,x)
    If =  1/2*np.log(3)
    trapErr = It - If
    simpError = Is- If
    diffTrapSimp=It-Is

    print(f"""N: {N} - Trapezregel: {It} - Fehler Trapezregel: {trapErr} -
         Simpsonregel: {Is} - Fehler Simpsonregel: {simpError} - Differenz Trapezregel-Simpsonregel: {diffTrapSimp}""")

    if(trapErr < math.pow(10,-8) and simpError < math.pow(10,-8)): #Bricht ab, falls Fehler klein genug werden
        break


```
