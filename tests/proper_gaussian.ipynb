```python
import numpy as np
```

```python
def lagrange_nodes() -> np.array:
    x1 = (1/3)*np.sqrt(5 + 2*np.sqrt(10/7))
    x2 = (1/3)*np.sqrt(5 - 2*np.sqrt(10/7))
    x3 = 0
    return np.array([-x1, -x2, x3, x2, x1])


def lagrange_weights() -> np.array:
    a3 = 128/225
    a2 = (322 + 13*np.sqrt(70))/900
    a1 = (322 - 13*np.sqrt(70))/900
    return np.array([a1, a2, a3, a2, a1])


x = lagrange_nodes()
a = lagrange_weights()
print(x)
print(a)
```

```python
def testfunction(x: np.array) -> np.array:
    return (x-0.2)**9


def gaussian_quadrature(xn: np.array, a: np.array):
    n = x.size
    yn = testfunction(xn)
    integr = 0.0

    for i in range(n):
        integr += a[i]*yn[i]
    return integr


xn = lagrange_nodes()
a = lagrange_weights()
integr = gaussian_quadrature(xn, a)
real_value = -0.608436224
print(integr)
print(np.abs(real_value - integr))
```
