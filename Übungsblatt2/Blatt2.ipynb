a)
Lineares Gleichgungssystem:

$\begin{pmatrix}
1.2969 & 0.8648\\
0.2161 & 0.1441
\end{pmatrix}
* 
\begin{pmatrix}
x_{1}\\
x_{2}
\end{pmatrix}
=
\begin{pmatrix}
0.8642\\
0.1440
\end{pmatrix}$



```python
import numpy as np
import math
import matplotlib.pyplot as plt

#np.linalg.solve(a.b) takes inputs a and b, returns x
#https://numpy.org/doc/stable/reference/generated/numpy.linalg.solve.html
#Ax=b

A = [[1.2969,0.8648],[0.2161,0.1441]]
b=[0.8642,0.1440]
x = np.linalg.solve(A,b)
print("Lösung für das LGS: ", x)
print("x1=",x[0]," und x2=",x[1])
print("Man beachte die unterschiedlichen Ausgaben :o")
```

```python
#np.random.normal draw random samples from normal distribution
#im gegensatz zum vergeschlagenen randn kann man bei normal loc und scale, also erwartungwert und standardabweichung direkt eingeben
#Die Funktion gibt direkt ein Array mit 20 einträgen aus
noise = np.random.normal(loc=0,scale=0.0001,size=20)

#erstellung eines arrays welche 20 x' einträge halten kann
x_noisy = np.zeros((20,2))

for i in range(20):
    #addition von b mit dem vorher erstellten noise. Ich hoffe es ist so gewollt, dass beide Einträge in b mit dem selben noise wert addiert werden :)
    b_noisy=b+noise[i]

    #wir nutzen den eben erstellten wert b' um das entsprechende x' zu berechnen und speichern diese in x_noisy
    x_noisy[i]=np.linalg.solve(A,b_noisy)


#Wir waren leider etwas von der Aufgabenstellung verwirrt, ob mit "Die Lösungen"
#jeweils x1 und x2 gemeint sind, oder tatsächlich das "gesamte" x, welches ja aber auch einfach nur ein Array mit 2 Einträgen ist.
#Daher haben wir uns entschieden beides zu berechnen und auszugeben.
#x_ges verhält sich hierbei logischerweise jeweils wie wenn man x1 und x2 zu einem Array mergt und dann den Durchschnitt und die Standardabweichung berechnet.

#Berechnung des Druchschnitts von x1 und x2 und x_ges
x1_mean = np.mean(x_noisy[:,0]) 
x2_mean = np.mean(x_noisy[:,1])
x_ges_mean = np.mean(x_noisy)

#Berechnung der Standardabweichung von x1 und x2 und x_ges
x1_std = np.std(x_noisy[:,0]) 
x2_std = np.std(x_noisy[:,1])
x_ges_std = np.std(x_noisy)

print("Gestörte Lösungen:\n",x_noisy)
print("Durchschnittliche Lösung von x1: ",x1_mean)
print("Durchschnittliche Lösung von x2: ",x2_mean)
print("Durchschnittliche Lösubg von x_ges: ",x_ges_mean)
print("Standardabweichung von x1: ",x1_std)
print("Standardabweichung von x2: ",x2_std)
print("Standardabweichung von x_ges: ",x_ges_std)

```

```python
def l2(x,x_noisy):
    #Variablenzuweisung für bessere lesbarkeit in Formel
    x1=x[0]
    x2=x[1]
    xn1=x_noisy[0]
    xn2=x_noisy[1]
    #Formel für L2 Fehler von Blatt
    return math.sqrt((xn1-x1)**2+(xn2-x2)**2)

def norm(x):
    #Formel für die Norm von x
    return math.sqrt((x[0]**2)+(x[1]**2))

def e_rel(x,x_l2_err):
    #Formel für den relativen Fehler von Blatt
    x_norm = norm(x)
    return x_l2_err/x_norm

x_l2 = np.zeros(20)
x_rel = np.zeros(20)

for i in range(20):
    x_l2[i]=l2(x,x_noisy[i])
    x_rel[i]=e_rel(x,x_l2[i])

print("L2 Fehler: ",x_l2)
print("Relative Fehler: ", x_rel)
```

d)

```python
#Berechnung des durchschnittlichen relativen Fehlers und der Standardabweichung des relativen Fehlers
rel_mean=np.mean(x_rel)
rel_std=np.std(x_rel)
print("Durchschnittlicher relativer Fehler: ",rel_mean)
print("Standardabweichung des relativen Fehlers: ", rel_std)

#Plotten des Fehlers als Boxplots
plt.figure(figsize=(8, 6))
plt.boxplot(x_rel)
plt.title('Boxplot des relativen Fehlers')
plt.ylabel('Relativer Fehler')
plt.show()
```
