```python
def f(x):
    return math.sinh(x)

def zweiterDiff(h,x):
    return (f(x+h)+(2*f(x))+f(x-h))/math.pow(h,2)

#Define x and h
x=0.6
exakt=math.sinh(x)

h=np.zeros(20)

for i in range(1,21):
    h[(i-1)]=math.pow(2,-i)

error = np.zeros(20)
for i in range(20):
    ah = zweiterDiff(h[i],0.6)
    error[i]=ah-exakt

print(error)


# plt.plot(range(1, 21), error, marker='o')
# plt.xlabel('Index of h')
# plt.ylabel('Error')
# plt.title('Error vs Index of h')
# plt.grid(True)
# plt.show()
```

```python
def lagrange2(x,x0,y0,x1,y1):
    l0 = (x-x1)/(x0-x1)
    l1=(x-x0)/(x1-x0)
    return y0*l0+y1*l1


def RichardsonExtrapolation2(x,h):
    errors = np.zeros(10)
    for i in range(len(h)):
        h0=h[i]
        h1=h[i]/2
        y0=zweiterDiff(h0,x)
        y1=zweiterDiff(h1,x)
        approx = lagrange2(x,h0,y0,h1,y1)
        errors[i]=np.abs(approx-exakt)
    return errors

#neues h, x und exakt so wie oben
h = np.zeros(10)

for i in range(1,11):
    h[i-1]=math.pow(2,-i)

rich2_erros = RichardsonExtrapolation2(x,h)
print(rich2_erros)
```

Wie man sieht wird unser Fehler immer größer. Das sollte bestimmt eigentlich nicht so sein, aber in Aufgabe a) wird der Fehler ja auch immer größer.
Da wird irgendwas nicht ganz richitg gelaufen sein, aber wir sind dann doch ein bisschem zu blöd gewesen zu finden was nicht richtig lief.

```python
def lagrange3(x,x0,y0,x1,y1,x2,y2):
    l0=((x-x1)*(x-x2))/((x0-x1)*(x0-x2))
    l1=((x-x0)*(x-x2))/((x1-x0)*(x1-x2))
    l2=((x-x0)*(x-x1))/((x2-x0)*(x2-x1))
    return y0*l0+y1*l1+y2*l2

def RichardsonExtrapolation3(x,h):
    errors = np.zeros(10)
    for i in range(len(h)):
        h0=h[i]
        h1=h[i]/2
        h2=h[i]/4
        y0=zweiterDiff(h0,x)
        y1=zweiterDiff(h1,x)
        y2=zweiterDiff(h2,x)
        approx = lagrange3(x,h0,y0,h1,y1,h2,y2)
        errors[i]=np.abs(approx-exakt)
    return errors

#h wie aus aufgabe c), x und exakt wie in a) definiert
rich3_errors = RichardsonExtrapolation3(x,h)
print(rich3_errors)
```

Wie auch schon in Teil c) wird es hier immer ungenauer.

```python
def h2Richardson(x,h):
    errors = np.zeros(10)
    for i in range(len(h)):
        h0=h[i]
        h1=h[i]/2
        x0=math.pow(h0,2)
        y0=zweiterDiff(h0,x)
        x1=math.pow(h1,2)
        y1=zweiterDiff(h1,x)
        approx = lagrange2(x,x0,y0,x1,y1)
        errors[i]=np.abs(approx-exakt)
    return errors

#x und h wie in c)
h2_rich_errors = h2Richardson(x,h)
print(h2_rich_errors)
```
