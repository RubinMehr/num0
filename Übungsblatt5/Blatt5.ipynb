```python
def f(x):
    return math.sinh(x)

def zweiterDiff(h,x):
    return (f(x+h)+(2*f(x))+f(x-h))/(h**2)

#Define x and h
x=0.6
exakt=math.sinh(x)

h=np.zeros(20)

for i in range(1,21):
    h[(i-1)]=math.pow(2,-i)

error = np.zeros(20)
for i in range(20):
    ah = zweiterDiff(h[i],0.6)
    error[i]=ah-exakt

print(error)


plt.plot(range(1, 21), error, marker='o')
plt.xlabel('Index of h')
plt.ylabel('Error')
plt.title('Error vs Index of h')
plt.grid(True)
plt.show()
```
